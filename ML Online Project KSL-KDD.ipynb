# Combining Train and Test Datasets
we observed that in train and test data we have different number of unique elements
 so, we will combine the train and test data and  wewill do preprocessing.
 and after preprocessing we will again separete data

```python
df = pd.concat([train, test], axis=0, ignore_index=True)
```

```python
df.shape
```

```python
df.describe(include = 'O')
```

```python
df.attack.unique() # if attack(nornal)  = good else bad
```

If attack type is **nornal** then its **good** condition otherwise **bad** so w'll do **binary classification**
but there are also different categories of attacks as below:

NO |Attack Type  | Attack Name
:--|:------------|:------------
1|**DOS**          | Back, land, Neptune, pod, smurf, teardrop
2|**User to Root Attacks**|Buffor_overflow, loadmodule, perl, rootkit
3|**Remote to Local Attacks**|ftp_write, guess_passwd, imap, multihop, phf, spy, warezclient
4|**Probes**|Satan, ipsweep, nmap, portsweep

```python
# this function is used to map attack types as above table

def attack_type(attack):
    if attack in ['back', 'land', 'neptune', 'pod', 'smurf', 'teardrop']:
        return 1
    elif attack in ['buffor_overflow', 'loadmodule', 'perl', 'rootkit']:
        return 2
    elif attack in ['ftp_write', 'guess_passwd', 'imap', 'multihop', 'phf', 'spy', 'warezclient']:
        return 3
    elif attack in ['satan', 'ipsweep', 'nmap', 'portsweep']:
        return 4
    else:
        return 0
```

```python
# we created new feature for attack [0,1,2,3,4] multiclass-classification
df['attack_multi_class'] = df.attack.apply(attack_type)
```

```python
# we created new feature for binary classifiaction 
df['attack_binary_class'] = df.attack.map(lambda x:0 if x=='normal' else 1)
```

```python
# this function is used to visualise the discrete columns
def discrete_plot(df, df1, feature):
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1,2,1)
    sns.countplot(x=feature, data=df)
    
    plt.subplot(1,2,2)
    sns.countplot(x=feature, data=df1)
    
    plt.show()

# this function is used to visualise the continuous columns
def continuous_plot(df, df1, feature):
    plt.figure(figsize=(15, 5))
    plt.subplot(1,2,1)
    sns.distplot(df[feature])
    
    plt.subplot(1,2,2)
    sns.distplot(df1[feature])
    plt.show()
    
```

# Import Classification Algorithams

```python
from sklearn.linear_model import LogisticRegression   #1
from sklearn.linear_model import SGDClassifier        #2
from sklearn.neighbors import KNeighborsClassifier    #3
from sklearn.tree import DecisionTreeClassifier       #4
from sklearn.ensemble import RandomForestClassifier   #5
from sklearn.naive_bayes import GaussianNB            #6
from sklearn.svm import SVC                           #7
```

# Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# creating pipelines for each classifire

# Logistic Regressoin
pipeline_lr = Pipeline([('scaler_lr', StandardScaler()),
                      ('pca_lr', PCA(n_components=3)),
                      ('model_lr', LogisticRegression(random_state=0))])

# SGDClassifire
pipeline_sgd = Pipeline([('scaler_sgd', StandardScaler()),
                      ('pca_sgd', PCA(n_components=3)),
                      ('model_sgd', SGDClassifier())])

# KNN
pipeline_knn = Pipeline([('scaler_knn', StandardScaler()),
                      ('pca_knn', PCA(n_components=3)),
                      ('model_knn', KNeighborsClassifier(n_neighbors=355))])
                        
# DecisionTree Classifier
pipeline_dt = Pipeline([('scaler_dt', StandardScaler()),
                      ('pca_dt', PCA(n_components=3)),
                      ('model_dt', DecisionTreeClassifier())])

# RandomForest Classifier
pipeline_rf = Pipeline([('scaler_rf', StandardScaler()),
                      ('pca_rf', PCA(n_components=3)),
                      ('model_rf', RandomForestClassifier())])

# GaussianNB
pipeline_gnb = Pipeline([('scaler_gnb', StandardScaler()),
                      ('pca_gnb', PCA(n_components=3)),
                      ('model_gnb', GaussianNB())])

# SVC
pipeline_svc = Pipeline([('scaler_svc', StandardScaler()),
                      ('pca_svc', PCA(n_components=3)),
                      ('model_svc', SVC(kernel='linear'))])
```

```python
# creating list of pipelines
pipelines = [pipeline_lr, pipeline_sgd, pipeline_knn, pipeline_dt, pipeline_rf, pipeline_gnb, pipeline_svc]


pipe_dict = {0:'LogisticRegression',
             1:'SGDClassifier',
             2:'KNeighborsClassifier',
             3:'DecisionTreeClassifier',
             4:'RandomForestClassifier',
             5:'GaussianNB',
             6:'SVC'}

# list of classifire names
model_names = ['LogisticRegression',
             'SGDClassifier',
             'KNeighborsClassifier',
             'DecisionTreeClassifier',
             'RandomForestClassifier',
             'GaussianNB',
             'SVC']
```

```python
# binary classification
# traing each model with pipeline

for pipe in pipelines:
    pipe.fit(XX_train, y_train_binary_final)
```

```python
# finding score of train and test data

train_acc = []
test_acc = []
for pipe_pred in pipelines:
    train_acc.append(pipe_pred.score(XX_train, y_train_binary_final))
    test_acc.append(pipe_pred.score(XX_test, y_test_binary_final))
```

```python
# creating dataframe for train and test data with thier acuuracy scores 
accuracies = pd.DataFrame({'Model Name':model_names, 'Train Accuracy': train_acc, 'Test Accuracy':test_acc})
```

```python
accuracies
```

```python
# plotting line plot of traing vs test

plt.figure(figsize=(10, 5))
sns.set_style('darkgrid')
sns.lineplot(accuracies['Model Name'], accuracies['Train Accuracy'], marker='o', label='Traing')
sns.lineplot(accuracies['Model Name'], accuracies['Test Accuracy'], marker='o', label='Testing')

plt.xticks(rotation=45)
plt.show()
```
